```python
import numpy as np
import matplotlib.pyplot as pl
import matplotlib.lines as ln

data_file = open('ISSN_D_tot.csv', 'r')

data_array = np.zeros((72104,), dtype=[('year', int), ('month', int), ('day', int), ('date',float), ('sunspots', int), ('official', int)])

index = 0;
max_index = 0;

for line in data_file:
    (year, month, day, date, sunspots, official) = line.split(',')
    data_array[index]=(int(year), int(month), int(day), float(date), int(sunspots), (official))
    
    if(float(date) > float(data_array[max_index]['date'])):
        max_index = index
    
    index += 1

data_file.close()
    
print("The most recent sunspot observations were on %d-%d-%d" % (data_array[max_index]['year'], data_array[max_index]['month'], data_array[max_index]['day']))
print("Statistics of daily sunspot observations (1818-present)")
print("Mean:", np.mean(data_array[:]['sunspots'][data_array[:]['sunspots']!=-1]))
print("Median:", np.median(data_array[:]['sunspots'][data_array[:]['sunspots']!=-1]))
print("Mode:", np.std(data_array[:]['sunspots'][data_array[:]['sunspots']!=-1]))


pl.scatter(data_array[:]['date'], data_array[:]['sunspots'], c='cyan')
```
